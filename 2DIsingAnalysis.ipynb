```python
# Import Statements
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# X-axis
T = np.arange(2,3,step=0.02)
```

```python
# Func to read in data files
def read_data(fileName):
    df = pd.read_csv(fileName)
    df = pd.read_csv(fileName,header=None)
    dataArr = df.to_numpy()
    dataArr = dataArr[0]
    dataArr = np.delete(dataArr, 50)
    return dataArr
```

```python
# Creating data arrays (from text files created by .cpp file)
M1_L90 = read_data('M1-L90.txt')
M2_L90 = read_data('M2-L90.txt')
M4_L90 = read_data('M4-L90.txt')
E1_L90 = read_data('E1-L90.txt')
E2_L90 = read_data('E2-L90.txt')

M1_L60 = read_data('M1-L60.txt')
M2_L60 = read_data('M2-L60.txt')
M4_L60 = read_data('M4-L60.txt')
E1_L60 = read_data('E1-L60.txt')
E2_L60 = read_data('E2-L60.txt')

M1_L30 = read_data('M1-L30.txt')
M2_L30 = read_data('M2-L30.txt')
M4_L30 = read_data('M4-L30.txt')
E1_L30 = read_data('E1-L30.txt')
E2_L30 = read_data('E2-L30.txt')
```

```python
# Calculating Quantities from Data
def Magnetization(M1,L):
    return np.abs(M1) / (L**2)
def SpecificHeat(E2,E1,L):
    return ((E2) - ((E1)**2)) / ((T**2)*(L**2))
def Susceptibility(M2,M1,L):
    return ((M2) - (np.abs(M1))**2) / ((T**2)*(L**2))
def B4(M4,M2,L):
    return 1 - (M4/(3*(M2)**2))
```

```python
beta = 1/T
# Critical Temperature
Tc = 2/np.log(1 + np.sqrt(2))
# Analytic Magnetization Function
A_Magnetization = np.zeros(50)
A_Magnetization[0:12] = np.power(1 - np.power(np.sinh(2*beta[0:12]), -4),1/8)
A_Magnetization[12:50] = 0
```

```python
# All 4 plots in 1... for fun!
fig, axs = plt.subplots(2,2,figsize=(10,10))
# Magnetization
axs[0,0].set_title('Magnetization')
axs[0,0].plot(T,Magnetization(M1_L90,90),color='olive',label='L = 90')
axs[0,0].scatter(T,Magnetization(M1_L90,90),s=20,alpha=1,facecolors='none',edgecolors='olive')
axs[0,0].plot(T,Magnetization(M1_L60,60),color='navy',label='L = 60')
axs[0,0].scatter(T,Magnetization(M1_L60,60),s=20,alpha=1,facecolors='none',edgecolors='navy')
axs[0,0].plot(T,Magnetization(M1_L30,30),color='peru',label='L = 30')
axs[0,0].scatter(T,Magnetization(M1_L30,30),s=20,alpha=1,facecolors='none',edgecolors='peru')
axs[0,0].set_xlim(2,2.5)
axs[0,0].grid(alpha=0.5)
axs[0,0].set_ylabel('m',size=15,rotation='horizontal')
axs[0,0].set_xlabel('T',size=15)
axs[0,0].legend(framealpha=0)
# Susceptibility
axs[0,1].plot(T,Susceptibility(M2_L90,M1_L90,90),color='olive',label='L = 90')
axs[0,1].scatter(T,Susceptibility(M2_L90,M1_L90,90),s=20,alpha=1,facecolors='none',edgecolors='olive')
axs[0,1].plot(T,Susceptibility(M2_L60,M1_L60,60),color='navy',label='L = 60')
axs[0,1].scatter(T,Susceptibility(M2_L60,M1_L60,60),s=20,alpha=1,facecolors='none',edgecolors='navy')
axs[0,1].plot(T,Susceptibility(M2_L30,M1_L30,30),color='peru',label='L = 30')
axs[0,1].scatter(T,Susceptibility(M2_L30,M1_L30,30),s=20,alpha=1,facecolors='none',edgecolors='peru')
axs[0,1].set_xlim(2,2.5)
axs[0,1].grid(alpha=0.5)
axs[0,1].set_xlabel('T',size=15)
axs[0,1].set_ylabel(r'$\chi$',size=15,rotation='horizontal')
axs[0,1].set_title('Magnetic Susceptibility')
axs[0,1].legend(framealpha=0)
# Specific Heat
axs[1,0].plot(T,SpecificHeat(E2_L90,E1_L90,90),color='olive',label='L = 90')
axs[1,0].scatter(T,SpecificHeat(E2_L90,E1_L90,90),s=20,alpha=1,facecolors='none',edgecolors='olive')
axs[1,0].plot(T,SpecificHeat(E2_L60,E1_L60,60),color='navy',label='L = 60')
axs[1,0].scatter(T,SpecificHeat(E2_L60,E1_L60,60),s=20,alpha=1,facecolors='none',edgecolors='navy')
axs[1,0].plot(T,SpecificHeat(E2_L30,E1_L30,30),color='peru',label='L = 30')
axs[1,0].scatter(T,SpecificHeat(E2_L30,E1_L30,30),s=20,alpha=1,facecolors='none',edgecolors='peru')
axs[1,0].grid(alpha=0.5)
axs[1,0].set_xticks([2,2.2,2.4,2.6,2.8])
axs[1,0].set_yticks([0.5,1,1.5,2])
axs[1,0].set_title('Specific Heat')
axs[1,0].set_ylabel('c',size=15,rotation='horizontal')
axs[1,0].set_xlabel('T',size=15)
axs[1,0].legend(framealpha=0)
# Binder Cumulant
axs[1,1].axvline(Tc,color='black',linestyle='dashed',alpha=0.8)
axs[1,1].plot(T,B4(M4_L90,M2_L90,90),color='olive',label='L = 90')
axs[1,1].scatter(T,B4(M4_L90,M2_L90,90),s=20,alpha=1,facecolors='none',edgecolors='olive')
axs[1,1].plot(T,B4(M4_L60,M2_L60,60),color='navy',label='L = 60')
axs[1,1].scatter(T,B4(M4_L60,M2_L60,60),s=20,alpha=1,facecolors='none',edgecolors='navy')
axs[1,1].plot(T,B4(M4_L30,M2_L30,30),color='peru',label='L = 30')
axs[1,1].scatter(T,B4(M4_L30,M2_L30,30),s=20,alpha=1,facecolors='none',edgecolors='peru')
axs[1,1].set_title('Binder Cumulant')
axs[1,1].set_xlim(2.1,2.4)
axs[1,1].set_ylim(0.2,0.75)
axs[1,1].set_yticks([0.65,0.55,0.45,0.35,0.25])
axs[1,1].grid(alpha=0.5)
axs[1,1].text(2.28,0.6175,r'$T_c \approx 2.269$',size=15)
axs[1,1].legend(framealpha=0)
axs[1,1].set_xlabel('T',size=15)
axs[1,1].set_ylabel(r'$B_4$',size=15,rotation='horizontal')
plt.tight_layout()
# Save Plot
plt.tight_layout()
plt.savefig('2DIsing.png',bbox_inches='tight')
plt.show()
```
